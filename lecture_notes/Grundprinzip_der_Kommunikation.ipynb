# Grundprinzip der Kommunikation

Wir wissen bereits das eine API die Kommunikation zwischen verschiedenen Softwareanwendungen ermöglicht.
Unter verschiedenen Softwareanwendungen, kann man sich Anwendungen vorstellen, welche in unterschiedlichen Technologien entwickelt wurden. Nun stellt sich natürlich die Frage, wie genau diese Art von Kommunikation, zwischen unterschiedlichen Anwendungen funktioniert.
<br>
<br>
Die grundlegende Theorie hinter APIs basiert auf dem Prinzip der Kommunikation zwischen den beiden Komponenten, Client und dem Server. Man spricht in diesem Fall vom sogenannten Client-Server-Modell. Es ist eine Architektur, bei der zwei separate Softwarekomponenten – der Client und der Server – miteinander interagieren. Sie haben unterschiedliche Rollen:
- Der Client ist derjenige, der eine Anfrage (Request) stellt. Es handelt sich dabei in der Regel um eine Anwendung oder ein Gerät, das vom Benutzer gesteuert wird (z. B. ein Webbrowser, eine mobile App oder eine andere Software). Mit Anfrage meint man eine Anforderung oder Abfrage, um bestimmte Informationen oder eine bestimmte Aktion auszuführen.
- Der Server ist der Dienstanbieter, der die angeforderten Informationen oder Funktionen bereitstellt. Der Server verarbeitet die Anfrage des Clients, führt eine entsprechende Aktion aus (z. B. Daten aus einer Datenbank abrufen oder Berechnungen durchführen) und sendet eine Antwort (Response) zurück.

Bildlich kann man sich diesen Prozess folgendermaßen vorstellen:

<img src="../img/Client-Server_Modell_01.png" alt="Client-Server_Modell_01" width="300">

Grob gesagt läuft die Kommunikation in folgenden Schritten ab:

1. Der Client sendet eine Anfrage:
   - Der Client fordert über die API Informationen oder Funktionen vom Server an. Diese Anfrage kann durch die Interaktion des Benutzers mit einer Anwendung ausgelöst werden. Zum Beispiel könnte der Benutzer auf einer Wetter-App auf "Wetter abrufen" klicken.
2. Der Server empfängt die Anfrage und verarbeitet sie:
    - Der Server empfängt die Anfrage des Clients und prüft, was genau angefordert wurde. Der Server ist dafür verantwortlich, diese Anfrage zu verarbeiten, z. B. durch den Abruf von Daten aus einer Datenbank oder durch das Ausführen einer Berechnung.
3. Der Server sendet die Antwort zurück:
    - Nachdem der Server die Anfrage verarbeitet hat, wird eine Antwort zurück an den Client gesendet. Diese Antwort enthält die angeforderten Daten oder eine Bestätigung der durchgeführten Aktion.
4. Der Client verarbeitet die Antwort und zeigt sie an:
    - Der Client erhält die Antwort und zeigt sie in einer für den Benutzer verständlichen Weise an, z. B. indem er die Wetterdaten auf einer Benutzeroberfläche darstellt.

So können wir unser Bild der Kommunikation weiter verfeinern:

Im Client-Server-Modell fungiert die API als Vermittler zwischen dem Client und dem Server. Sie definiert eine Reihe von Regeln und Protokollen, die beschreiben, wie der Client und der Server miteinander kommunizieren sollen. Diese Regeln umfassen:
- Wie der Client eine Anfrage stellen muss
- Wie der Server die Anfrage verarbeiten und eine Antwort zurückgeben muss
- Welches Format die Daten haben müssen

Die API sorgt also auch dafür, dass der Client eine strukturierte Anfrage stellen kann und der Server die Daten in einem für den Client verständlichen Format zurückgibt.
<br>
<br>
Mit Regeln und Protokollen meint man die vereinbarten Standards, Vorschriften und Verfahren, die festlegen, wie die Kommunikation zwischen dem Client und dem Server über die API ablaufen soll. Typische Regeln bei der Kommunikation könnten sein:

1. Formate und Strukturen:
   - Die API legt fest, wie die Daten strukturiert werden müssen, damit sie vom Server oder Client korrekt verarbeitet werden können. Häufig verwendete Formate sind JSON (JavaScript Object Notation) oder XML (Extensible Markup Language). Zum Beispiel könnte eine Regel besagen, dass alle Antworten in JSON formatiert sein müssen. Mit diesen Formaten werden wir uns noch im Detail auseinandersetzen.

2. Fehlerbehandlung:
   - APIs definieren Regeln darüber, wie Fehler behandelt und gemeldet werden. Eine häufige Regel könnte sein, dass im Falle eines Fehlers der Server eine Fehlerantwort mit einem entsprechenden HTTP-Statuscode (z.B 404) zurückgibt. Was HTTP genau ist und was man unter Statuscodes versteht wird später erklärt.

3. Zugangskontrolle:
   - APIs können Regeln für die Autorisierung und Authentifizierung enthalten, die festlegen, wie der Client nachweisen muss, dass er berechtigt ist, auf bestimmte Ressourcen zuzugreifen. Dies könnte durch API-Schlüssel, OAuth-Token oder JSON Web Tokens (JWT) geschehen. Die drei genannten Methoden zur Autorisierung und Authentifizierung werden ebenfalls im Details seperat betrachtet.
   - Bei der Authentifizierung wird geprüft, wer eine Person oder ein System ist. Es geht darum, die Identität des Benutzers oder Systems zu bestätigen, der oder das versucht, auf ein System oder eine Ressource zuzugreifen.  Die Authentifizierung kann zum Beispiel durch Benutzername und Password erfolgen.
   - Autorisierung ist der Prozess, bei dem überprüft wird, was eine authentifizierte Person oder ein System tun darf. Sie stellt sicher, dass ein Benutzer nur auf die Ressourcen zugreifen oder Aktionen ausführen kann, für die er die entsprechenden Berechtigungen hat.

4. Ratenbegrenzung:
   - Es können Regeln festgelegt werden, die steuern, wie oft der Client Anfragen an die API stellen darf. Eine Regel könnte z. B. besagen, dass der Client maximal 1000 Anfragen pro Stunde stellen darf.

Jetzt bleibt noch zu klären was genau ein Protokoll ist. Es ist ein vereinbartes Kommunikationsverfahren oder eine technische Spezifikation, die beschreibt, wie die Daten zwischen dem Client und dem Server ausgetauscht werden. Betrachtet die Protokolle als Regeln und Standards, welche den Datenaustausch zwischen Systemen über das Internet oder andere Netzwerke regeln. Zwei der weit verbreitesten Protokolle sind HTTP (Hypertext Transfer Protocol) und SOAP (Simple Object Access Protocol).
<br>
<br>
Das HTTP-Protokol ist das am häufigsten verwendete Protokoll für Web-APIs. HTTP definiert, wie Anfragen vom Client an den Server gesendet werden und wie Antworten vom Server zurück zum Client gelangen. Es beschreibt auch, wie Fehler behandelt werden und wie die Daten während der Kommunikation strukturiert sind. Jedoch sind solche Protokolle wie SOAP ebenfalls heutzutage nicht wegzudenken. Ich würde an dieser Stelle nicht in die Tiefe der Protokolle gehen, dazu werden wir noch kommen.
<br>
<br>
Mit diesem Wissen können wir wieder unser Bild vom Client-Server-Modell erweitern:

APIs werden häufig nach deren Architektur unterteilt. Architekturen beschäftigen sich mehr mit dem Thema wie eine API entworfen und implementiert wird. Kurz gesagt: Es ermöglicht die Festlegung, wie die Daten und Funktionen eines Softwaresystems strukturiert, organisiert und verfügbar gemacht werden. Für den Anfang reicht es aus zu wissen, welche API-Architekturen in der Praxis verwendet werden. Die bekanntesten Architekturen sind:
- REST-Architektur (Representational State Transfer): Eine Architektur, die auf HTTP basiert und Zustandslos ist. Was genau das bedeutet, wird in einem extra Kapitel behandelt.
- SOAP-Architektur: Eine Architektur welche auf dem SOAP Protokoll basiert mit hoher Sicherheit. Diese Architektur ist heuzutage immer noch wichtig und muss in einem seperaten Kapitel behandelt werden.

Es gibt viele Möglichkeiten APIs zu unterteilen:

1. Nach Zugriffskontrolle und Sichtbarkeit
   - Öffentliche APIs (Public APIs)
   - Private APIs
   - Partner-APIs
2. Nach Datenformat und Kommunikationsprotokollen
   - REST-APIs (Representational State Transfer)
   - SOAP-APIs (Simple Object Access Protocol)
   - GraphQL-APIs
   - gRPC-APIs (gRPC Remote Procedure Call)
3. Nach Funktionsweise der API
   - RESTful APIs
   - WebSocket-APIs
   - RPC-APIs (Remote Procedure Call)
4. Nach Verwendungszweck und Funktion
   - Daten-APIs
   - Dienst-APIs
   - Integrations-APIs
   - Web-APIs
5. Nach Kommunikationsmuster
   - Synchron (Blocking) APIs
   - Asynchron (Non-blocking) APIs

Wie ihr sehen könnt, wird man praktisch mit enorm viel Informationen überschüttelt. Die Unterteilung der APIs hat kein Ende und man kann schnell den Überblick verlieren. Am Anfang sollte man sich auf die REST-APIs konzentrieren, damit ist man in der Praxis am häufigsten konfrontiert.

## Beispiel einer API

Betrachten wie eine einfache kostenlose API, welche uns den täglichen UV-Index eines Ortes angibt. Um diese API zu verwenden muss die folgende URL z.B. im Internet-Browser aufgerufen werden: <br>
https://api.open-meteo.com/v1/forecast?latitude=52.52&longitude=13.41&daily=uv_index_max
<br>
<br>
So sieht die Antwort, auch als Respone bezeichnet, im Browser dan naus:

<img src="../img/API_Example.png" alt="API_Example" width="500">

In dem Response der API können wir unterschiedliche Informationen entnehmen:
- Als erstes sehen wir die geografischen Daten, zum einen Latitude und Longitude. Bei dem Standort handelt es sich um Berlin.
- Auch Zeitinformationen sind vorhanden. Wir sehen die Verarbeitungszeit in Millisekunden, welche die API benötigt hatte um die Anfrage (im englsichen Request) zu verarbeiten. Die Zeitzone ist ebenfalls angegeben, es handelt sich um GMT. Es sind noch andere zeitliche Informationen vorhanden, man kann sie gerne recherchieren.
- Das interessanteste an der Response, sind die Wetterdaten. Wir sehen die 7 Tage mit Datum und der jeweilige UV-Index.

Diese API können wir in unserem Browser noch mehr erkunden, in dem wir die sogenannten Dev-Tools verwenden. Es handelt sich um eine Sammlung von Tools und Funktionen, die Entwicklern helfen, Webseiten zu analysieren, zu debuggen und zu optimieren. Für uns ist Netzwerk-Panel von Bedeutung, denn da finden wir alle Netzwerkaktivitäten und Anfragen die beim Laden einer Seite passieren.
<br>
<br>
Nachdem die Dev-Tools geöffnet wurden, klickt man auf das Netzwerk-Panel. In der linken Seitenleiste, findet man eine Datei mit einem ähnlichen Namen wie `forecast?latitude=52.52&longitude=13.41&daily=uv_index_max`. Wenn man diese Datei öffnen und dann auf das Respone-Panel klickt, sieht man die Antwort der API als JSON-Objekt dargestellt:

<img src="../img/API_Dev_Tools.png" alt="API_Example" width="500">

Sehr viele APIs geben eine Response in Form von JSON-Objekten zurück. Es handelt sich um eine standardisierte Datenstruktur, die verwendet wird, um Daten in einem leicht verständlichen und strukturierten Format zu speichern und auszutauschen. Es ist einfach zu lesen und zu schreiben und kann in den meisten modernen Programmiersprachen problemlos verarbeitet werden.
<br>
<br>
Wenn wir zu dem Payload-Panel wechseln, dann sehen wir den Payload von unserem Request. Mit dem Payload werden hier die Daten bezeichnet, welche wir bei dem Request an die API mitgeben. Es ist also der eigentliche Inhalt der Kommunikation, welcher an den Server übertragen wird. In unserem Fall fordern wir den Server auf, uns UV-Index Informationen zu einem bestimmten Standort zu geben.
<br>
<br>
Betrachten wir nun den Headers-Panel, hier sind ebenfalls viele interessante Informationen vorhanden.

<img src="../img/API_Example_Header.png" alt="API_Example_Header" width="600">

Unter dem Abschnitt "General" findet man allgemeine Informationen zur Kommunikation. Zum einen sieht man die verwendete Request Methode, es handelt sich um eine HTTP-GET-Methode. Wir werden uns noch mit den HTTP-Methoden ausführlich beschäftigen. Man sieht hier noch den sogenannten Status Code 200, er steht für eine erfolgreiche Anfrage. Es sind viele unterschiedliche Status Codes vorhanden, wir werden uns sie alle im Detail anschauen.
<br>
<br>
Man findet noch Informationen zu dem Response-Header und dem Request-Header. Ein Header ist ein Bereich, der zusätzliche Informationen über die Response oder die Request enthält. Header bestehen aus Schlüssel-Wert-Paaren und helfen dabei, die Kommunikation zwischen einem Client (z. B. Webbrowser) und einem Server zu steuern und zu definieren, wie die Daten verarbeitet und übertragen werden. Schauen wir uns einfach mal ein paar dieser Informationen beim Response Header an.

```txt
HTTP/1.1 200 OK
Date: Mon, 30 Dec 2024 10:20:32 GMT
Content-Type: application/json; charset=utf-8
Transfer-Encoding: chunked
Connection: keep-alive
Content-Encoding: deflate
```

- `HTTP/1.1` zeigt an, dass die Antwort nach der Version 1.1 des HTTP-Protokolls gesendet wird.
- `200 OK` bedeutet, dass die Anfrage erfolgreich verarbeitet wurde und der Server die angeforderten Daten zurückgibt. Dies ist der Standardstatuscode für eine erfolgreiche Anfrage.
- `Date: Mon, 30 Dec 2024 10:20:32 GMT` gibt an, wann die Antwort vom Server generiert wurde.
- `Content-Type: application/json; charset=utf-8` gibt an dass der Inhalt der Antwort im JSON-Format sich befindet und die die UTF-8-Zeichenkodierung verwendet.
- `Transfer-Encoding: chunked` gibt an, dass die Antwort in "Chunks" (Stücken) gesendet wird.
- `Connection: keep-alive` gibt an, dass die Verbindung zwischen dem Client und dem Server nach Abschluss der aktuellen Anfrage offen bleibt.
- `Content-Encoding: deflate` gibt an, dass der Inhalt der Antwort mit dem Deflate-Kompressionsverfahren komprimiert wurde,  um die Größe der übertragenen Daten zu verringern. 

Man muss auf jeden Fall nicht alle Informationen die ich jetzt angesprochen hatte verstehen. Es sollte nur kurz oberflächlich angesprochen werden, um einen groben Überblick zu schaffen. Der Request-Header hat mehr Informationen, darauf werden wird jedoch erstmal nicht eingehen, wie bereits gesagt, wir wollen erstmal einen groben Überblick schaffen.
<br>
<br>
Wie wir gesehen hatten mussten wir eine bestimmte URL verwenden um eine Request an die API zu senden:
<br>
https://api.open-meteo.com/v1/forecast?latitude=52.52&longitude=13.41&daily=uv_index_max

Betrachten wir die Bestandteile dieser URL genauer, denn hier sind wichtige Informationen enthalten:
- `https://` Dies ist das Protokoll und gibt an, dass die Anfrage über das sichere HTTP-Protokoll (HTTPS) gemacht wird.
- `api.open-meteo.com` Dies ist der Hostname der API, der den Server angibt, mit dem die Anfrage kommuniziert wird. In diesem Fall handelt es sich um den Server von Open-Meteo, einem Anbieter von Wettervorhersagedaten.
- `/v1/forecast` Dies ist der API-Endpunkt. Der Endpunkt `/forecast` zeigt an, dass es sich um eine Vorhersageabfrage handelt. Die Version `v1` gibt die API-Version an, die verwendet wird. Die API-Endpunkte werden in Zukunft für uns noch sehr interessant sein, wir werden sie nicht nur verwenden, sondern auch selber erstellen.
- In der URL befinden sich noch sogenannte Abfrageparameter (Query-Parameter). Nach dem `?` folgt eine Liste von Abfrageparametern. Diese Parameter geben an, welche spezifischen Daten die API zurückgeben soll. Jeder Parameter besteht aus einem Schlüssel-Wert-Paar, das durch das Zeichen `=` verbunden ist, und mehrere Parameter sind durch `&` getrennt.
  - latitude=52.52
  - longitude=13.41
  - daily=uv_index_max

Im laufe der Zeit werden wir selber solche URLs erstellen und unterschiedliche Parameter definieren. Für den Anfang reicht es aus, zu wissen wie eine URL grob aufgebaut ist.

## URL und URI

Beim Zugriff auf API-Ressourcen stößt man irgendwann auf die beiden Begriffe URL und URI. Man sollte früh genug klären was genau unten den Begriffen zu verstehen ist und wo die Unterschiede liegen. Leider verwenden viel zu viele Menschen die beiden Begriffe als Synonyme.
<br>
<br>
Eine URI (Uniform Resource Identifier) ist ein allgemeiner Begriff für eine Zeichenkette, die eine Ressource eindeutig identifiziert. Somit handelt es sich um diese Zeichenkette: `https://api.open-meteo.com/v1/forecast?latitude=52.52&longitude=13.41&daily=uv_index_max` um eine URI, da sie eine bestimmte Ressource identifiziert.
<br>
<br>
Eine URL (Uniform Resource Locator) ist eine spezielle Art von URI, die zusätzlich Informationen zur Abrufmethode (Protokoll) und dem Zugriffsweg (Pfad + Parameter) enthält. Sie identifiziert also nicht nur die Ressource, sondern sagt auch aus, wo sie sich befindet und wie man drauf zugreift. 
Jede URL ist also eine URI, aber nicht jede URI ist eine URL! Zum Beispiel ist `/v1/forecast` eine URI, weil eine Ressource eindeutig identifiziert wird, aber es ist keine URL, denn das Protokoll und der Host fehlen.
<br>
<br>
Betrachten wir von der URI/URL `https://api.open-meteo.com/v1/forecast?latitude=52.52&longitude=13.41&daily=uv_index_max` die einzelnen Bestandteile genauer:

- **Schema (Protokoll):** <br>
  `https` gibt an, dass die Kommunikation über HTTPS (gesicherte Verbindung) erfolgt.

- **Host (Domain):** <br>
  `api.open-meteo.com` ist die Adresse, welche man benötigt, um eine Website zu besuchen

- **Pfad:** <br>
  `/v1/forecast` gibt an, welche Ressource oder Funktion aufgerufen wird (hier die Wettervorhersage der API in Version 1).

- **Query-Parameter:** <br>
  `?latitude=52.52&longitude=13.41&daily=uv_index_max` enthält zusätzliche Daten zur Anpassung der Anfrage.
