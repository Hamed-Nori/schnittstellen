# Das Zeitalter des Internets und die ersten Web-APIs

Die 1990er Jahre markierten einen Wendepunkt mit dem Aufkommen des World Wide Web. Plötzlich war es möglich, Anwendungen zu entwickeln, die über das Internet kommunizierten. Dies führte zur Entwicklung von Web-APIs, die es Web-Anwendungen ermöglichten, Daten und Dienste über das Internet auszutauschen. Eine Web-API ist wie eine gemeinsame Sprache, die festlegt, wie Anwendungen über das Internet miteinander sprechen können.
<br>
<br>
Ein Meilenstein in dieser Entwicklung war die Einführung von SOAP (Simple Object Access Protocol) im Jahr 1998. SOAP ist ein Satz von Regeln, der definiert, wie Nachrichten zwischen Anwendungen über das Internet gesendet werden. Es ermöglicht verschiedenen Anwendungen, auch wenn sie in unterschiedlichen Programmiersprachen geschrieben sind oder auf verschiedenen Systemen laufen, miteinander zu kommunizieren.
<br>
<br>
Stellt euch vor, man möchte mit jemandem in einem anderen Land sprechen, der eine andere Sprache spricht. Man könnte einen Übersetzer verwenden, der die Nachricht in eine gemeinsame Sprache übersetzt. In diesem Fall ist SOAP wie der Übersetzer für die Anwendungen.
<br>
<br>
SOAP nutzte XML zur Strukturierung von Nachrichten und ermöglichte es Anwendungen, komplexe Daten über HTTP zu senden und zu empfangen.
<br>
<br>
XML steht für Extensible Markup Language, es ist eine Sprache, die verwendet wird, um Daten in einer strukturierten und lesbaren Form darzustellen. Wenn man zum Beispiel eine Kontaktliste speichern möchte, könnte man XML verwenden. Dadurch kann man jeden Kontakt klar strukturieren und die Kontaktliste könnte zum Beispiel so aussehen:
```xml
<Kontakte>
    <Kontakt>
        <Name>Max Mustermann</Name>
        <Telefon>0123456789</Telefon>
    </Kontakt>
    <Kontakt>
        <Name>Erika Musterfrau</Name>
        <Telefon>0987654321</Telefon>
    </Kontakt>
</Kontakte>
```

Es ist korrekt, dass sowohl SOAP als auch HTTP Protokolle für die Kommunikation verwendet werden. Allerdings operieren sie auf unterschiedlichen Ebenen und erfüllen verschiedene Aufgaben innerhalb der Kommunikation zwischen Anwendungen. Tatsächlich werden SOAP und HTTP Protokolle oft gemeinsam verwendet, um den Datenaustausch über das Internet zu ermöglichen.
<br>
<br>
SOAP definiert die Struktur und den Inhalt von Nachrichten zwischen Anwendungen. Er legt also fest, wie Daten in einem standardisierten Format (XML) verpackt werden sollen, damit verschiedene Anwendungen sie verstehen können.
<br>
<br>
HTTP ist zuständig für die Übertragung von Daten über das Internet. Er bestimmt wie Datenpakete zwischen Client (z.B. Anwender) und Server (z.B. Anbieter) gesendet und empfangen werden.
<br>
<br>
So könnte die Kommunikation vereinfacht zwischen Client und Server aussehen:

1. **Verpacken:**
   Die Anwendung erstellt eine SOAP-Nachricht in XML, die die benötigten Daten und Anweisungen enthält.
   </br>
2. **Versand:**
   Diese Nachricht wird in eine **HTTP-Anfrage** eingebettet und an den Server gesendet.
   </br>
3. **Empfang:**
   Der Server empfängt die HTTP-Anfrage, entnimmt die SOAP-Nachricht und verarbeitet sie entsprechend.
   </br>
4. **Antwort:**
   Der Server kann ebenfalls eine SOAP-Nachricht in einer HTTP-Antwort zurücksenden.

**Konkretes Beispiel einer API (SOAP-Webdienst):**

Die Number Conversion Web Service API von Data Access Worldwide ist ein öffentlich zugänglicher SOAP-Webdienst, der einfache Funktionen zur Zahlkonvertierung bereitstellt. Diese API eignet sich hervorragend, um die Grundlagen von SOAP (Simple Object Access Protocol) zu erlernen, da sie leicht verständliche Funktionen bietet und über HTTP zugänglich ist.
<br>
<br>
Unter der folgenden URL kann man die API verwenden:
https://www.dataaccess.com/webservicesserver/NumberConversion.wso

Die API bietet zwei Hauptmethoden:

1. NumberToWords: Konvertiert eine Ganzzahl in ihre ausgeschriebene Wortform in englischer Sprache.
2. NumberToDollars: Konvertiert eine Dezimalzahl in die ausgeschriebene Wortform von US-Dollar.

Wir könnten an den Server eine SOAP-Anfrage senden um die Zahl `123` in die Wortform zu konvertieren. Der Server bekommt eine XML-Datei welche so aussehen könnte:
```xml
<?xml version="1.0" encoding="utf-8"?>
<soap:Envelope xmlns:xsi="http://www.w3.org/2001/XMLSchema-instance"
               xmlns:xsd="http://www.w3.org/2001/XMLSchema"
               xmlns:soap="http://schemas.xmlsoap.org/soap/envelope/">
  <soap:Body>
    <NumberToWords xmlns="http://www.dataaccess.com/webservicesserver/">
      <ubiNum>123</ubiNum>
    </NumberToWords>
  </soap:Body>
</soap:Envelope>
```

Wir müssen momentan nicht jede einzelne Zeile dieser XML-Datei verstehen, jedoch finden wir definitiv die Zeile mit der Zahl `123` wieder.

Nachdem der Server von uns (dem Client) eine Anfrage erhalten hat, wird er eine SOAP-Antwort an uns senden. Diese ist ebenfalls in einer XML-Datei und könnte so aussehen:
```xml
<?xml version="1.0" encoding="utf-8"?>
<soap:Envelope xmlns:soap="http://schemas.xmlsoap.org/soap/envelope/">
  <soap:Body>
    <NumberToWordsResponse xmlns="http://www.dataaccess.com/webservicesserver/">
      <NumberToWordsResult>one hundred twenty three</NumberToWordsResult>
    </NumberToWordsResponse>
  </soap:Body>
</soap:Envelope>
```

Irgendwo in dieser Datei sollte die folgende Zeile `one hundred twenty three` auffallen, welche die ausgeschriebene Zahl enthält. Später werden wir uns auch mit dieser API im Detail beschäftigen und den Aufbau der XML-Dateien verstehen.
