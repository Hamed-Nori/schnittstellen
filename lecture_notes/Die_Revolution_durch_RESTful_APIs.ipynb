# Die Revolution durch RESTful APIs

Im Jahr 2000 veröffentlichte Roy Fieldin seine Dissertation, in der er das REST (Representational State Transfer)-Architekturmodell vorstellte. 
<br>
<br>
REST basierte auf den bestehenden Standards des Webs und bot einen einfacheren, effizienteren und flexibleren Ansatz für den Datenaustausch zwischen Software Systemen im Vergleich zu vorherigen Methoden wie SOAP.
<br>
<br>
Außerdem definiert die REST Architektur eine Reihe Leitlinien, die dazu beitragen, skalierbare, leistungsfähige und wartbare Webdienste zu erstellen. Auf diese Leitlinien werden wir später erst eingehen, für das grobe Verständnis am Anfang, sind sie erstmal nicht wichtig.
<br>
<br>
Die RESTful APIs, also APIs welche auf der REST-Architektur aufgebaut sind, verwenden ebenfalls das HTTP Protokoll. Im Gegensatz zu SOAP ist die Art und Weise, wie das HTTP Protokoll verwendet wird, unterschiedlich.
<br>
<br>
Bei SOAP wird HTTP hauptsächlich als Transportmittel verwendet, um SOAP-Nachrichten zu senden. Die eigentliche  Bedeutung der Nachricht ist in der SOAP-Nachricht selbst (im XML-Format) verpackt.
<br>
<br>
Statt HTTP nur als Mittel zum Versenden von Nachrichten zu verwenden, nutzen REST-APIs die eingebauten Aktionen von HTTP direkt, um zu bestimmen, was mit den Daten passieren soll. Diese Aktionen sind wie Verben oder Befehle, die angeben, wie Daten verarbeitet oder sie bearbeitet werden sollen.
<br>
<br>
Ein weiterer wichtiger Unterschied ist die Verwendung unterschiedlicher Datenformate. Bei einer SOAP API wird hauptsächlich XML unterstützt. Bei einer REST API werden jedoch mehrere Datenformate unterstützt. Dazu zählen Formate wie z.B. JSON, YAML XML und noch mehr. Häufig werden JSON-Dateien wegen ihrer Leichtgewichtigkeit und hervorragenden Lesbarkeit bevorzugt. 
<br>
<br>
RESTful APIs sind aufgrund ihrer Einfachheit und der Nutzung von Webstandards zugänglicher und einfacher zu verwenden, selbst wenn man die zugrunde liegende REST-Architektur nicht vollständig versteht. SOAP-APIs hingegen erfordern ein detailliertes Verständnis des SOAP-Protokolls, um effektiv genutzt werden zu können.

**Konkretes Beispiel einer REST-API:**

Die JSONPlaceholder API ist eine ausgezeichnete Wahl für die demonstration einer REST API. Es ist ein kostenloser Online-REST-Dienst, der zum Testen und Prototyping von Anwendungen verwendet werden kann. Er stellt gefälschte Daten zur Verfügung, die in der Struktur echten API-Daten ähneln.
<br>
<br>
Unter der folgenden URL kann man die API verwenden:
https://jsonplaceholder.typicode.com/

Die API stellt folgende Ressourcen zur Verfügung:
- ⁠*Posts*: Beiträge
⁠- *Comments*: Kommentare
- ⁠*Albums*: Alben
- ⁠*Photos*: Fotos
- ⁠*Todos*: Aufgaben
- ⁠*Users*: Benutzer

Jede Ressource kann über eindeutige URLs (Endpoints) angesprochen werden.
<br>
<br>
Um die Einfachheit der Kommunikation mit einer RESTful API wie
JSONPlaceholder zu demonstrieren, zeige ich Ihnen eine sehr einfache Anfrage und die dazugehörige Datei.

1.⁠ ⁠Anfrage senden:
   
   Wir möchten einen "Beitrag" mit der "ID 1" abrufen. Die Anfrage geht an die folgede URL: ⁠ https://jsonplaceholder.typicode.com/posts/1 ⁠
   Diese Anfrage kann direkt in einem Webbrowser ausgeführt werden, indem Sie die URL in die Adressleiste eingeben.

2.⁠ ⁠Antowrt vom Server:
   Der Server antwortet mit einer *JSON-Datei*, die den angeforderten Beitrag enthält:
⁠ json
{
  "userId": 1,
  "id": 1,
  "title": "sunt aut facere repellat provident occaecati excepturi optio reprehenderit",
  "body": "quia et suscipit\nsuscipit recusandae consequuntur expedita et cum..."
}
 ⁠
