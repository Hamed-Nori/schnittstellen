# Die Anfänge der Software-Schnittstellen

In den 1940er und 1950er Jahren stand die Computertechnologie noch am Anfang. Programmierer arbeiteten direkt mit der Hardware, und es gab wenig bis keine Abstraktion. Mit der Zeit entstand jedoch der Bedarf, komplexe Systeme zu vereinfachen und die Entwicklung effizienter zu gestalten. Das Konzept der Schnittstelle wurde eingeführt, um die Kommunikation zwischen verschiedenen Teilen eines Programms oder zwischen unterschiedlichen Programmen zu ermöglichen.
<br>
<br>
Als Beispiel für die Computer und Programmierung in der Anfangszeit, ist der ENIAC ein hervorragendes Beispiel.
ENIAC steht für Electronic Numerical Integrator and Computer und gilt als einer der wichtigsten Computer in der Anfangszeit. 

<img src="../img/enica_01.jpg" alt="schnittstelle" width="300">

ENICA wurde ab 1942 im Auftrag von der US-Armee entwickelt und im Jahre 1946 der Öffentlichkeit vorgestellt. Die Hauptaufgabe des Computers war die Berechnung von ballistischen Tabellen. Der ENIAC konnte addieren, subtrahieren, multiplizieren, dividieren und Quadratwurzeln ziehen. Eine Addition/Subtraktion brauchte 0,2 Millisekunden, eine Multiplikation bis zu 2,8 ms, eine Division bis zu 24 ms und eine Quadratwurzel mehr als 300 ms.
<br>
<br>
Die Programmierung auf dem ENIAC war jedoch komplex und zeitaufwendig, da es keine Software im heutigen Sinne gab. Stattdessen musste der Rechner physisch umkonfiguriert werden, indem Kabel gesteckt, Schalter betätigt und Lochkarten genutzt wurden.

<img src="../img/eniac_02.jpg" alt="schnittstelle" width="300">

In den 1950er Jahren wurden in den USA die ersten drei weit verbreiteten, praktisch eingesetzten höheren Programmiersprachen entwickelt:
- Fortran 1954
- Lisp 1959
- COBOL 1959

Die vorgenannten Sprachen existieren mit ihren Nachfolgern bis heute. Vor allem LISP beeinflusste die später an amerikanischen Universitäten entwickelten Programmiersprachen stark.
<br>
<br>
In den Anfängen der Programmierung wurden Programme oft als ein großer Block von Code geschrieben. Dieser Ansatz führte jedoch zu mehreren Problemen:

⁠**Komplexität:**
  <br>
  Große Codeblöcke sind schwer zu verstehen, zu warten und zu debuggen.

**Wiederverwendbarkeit:** 
  <br>
  Ohne Unterteilung in kleinere Teile ist es schwierig, Codeabschnitte in anderen Teilen des Programms oder in anderen Projekten wiederzuverwenden.

**Teamarbeit:** 
  <br>
  Mehrere Entwickler können nicht effizient an demselben großen Codeblock arbeiten.

Um diese Probleme zu lösen, begannen Entwickler, ihre Programme in kleinere, logisch getrennte Einheiten zu unterteilen. Funktionen, Prozeduren oder Methoden sind grundlegende Bausteine in der Programmierung. Sie sind definierte Codeblöcke, die eine bestimmte Aufgabe erfüllen.

**Funktion:**
  <br>
  Ist ein Codeblock, der eine bestimmte Aufgabe ausführt, Eingabewerte (Parameter) entgegennimmt, eine Verarbeitung durchführt und oft einen Ausgabewert (Rückgabewert) liefert.

**Prozedur:**
  <br>
  Ist ein Codeblock, der eine Reihe von Anweisungen ausführt, aber im Gegensatz zu einer Funktion keinen Wert zurückgibt. Prozeduren führen Aktionen durch, ohne notwendigerweise ein Ergebnis zu liefern.

**Methode:**
  <br>
  Ist eine Funktion oder Prozedur, die innerhalb einer Klasse oder eines Objekts in der objektorientierten Programmierung (OOP) definiert ist. Methoden beschreiben das Verhalten von Objekten und können auf die internen Daten (Attribute) dieser Objekte zugreifen und sie verändern.

Die ersten Schnittstellen waren oft interne Methodenaufrufe oder Prozeduren innerhalb eines Systems. Sie dienten dazu, Funktionen zu kapseln und den Code modularer und wartbarer zu gestalten. Diese frühen Schnittstellen legten den Grundstein für das, was später als APIs bekannt wurde. 
<br>
<br>
Kapselung bedeutet in der Programmierung, die Implementierungsdetails einer Funktion vor dem Rest des Programms zu verbergen. Stattdessen wird eine definierte Schnitts
telle bereitgestellt, über die andere Teile des Programms mit dieser Funktion interagieren können.
